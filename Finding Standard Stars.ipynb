# Finding Standard Stars

### Imports

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import astropy.units as u
from astropy.coordinates import AltAz, EarthLocation, SkyCoord
from astropy.time import Time
import astroplan
from astroplan import FixedTarget
from astropy.io import fits
from astroquery.simbad import Simbad
customSimbad = Simbad()
customSimbad.add_votable_fields('flux(B)')
import warnings
warnings.filterwarnings("ignore")
```

### Create data table with right ascension, declination and B-mag for the northern-hemisphere standard stars

```python
#read in right ascension, declination values from landolt_north.csv
data = pd.read_csv("landolt_north.csv", delimiter = "|")
data.columns = data.columns.str.strip() #remove spacings in the headers
data = data.drop(['oid'], axis = 1) #no clue what the oid column is
data
#the next two parts sometimes take a minute to load
```

astroquery guide: https://astroquery.readthedocs.io/en/latest/

```python
#using SIMBAD, add another column with the magnitude of each standard star within the B-band 
data['b mag'] = None
for i in range (0, data.shape[0]):
    result_table = customSimbad.query_object(data['Identifier'].iloc[i])
    data['b mag'].iloc[i] = result_table[0][11]
data.head()
```

### Transform the right ascension, declination into altitude, azimuth co-ordinates

image header guide: https://docs.astropy.org/en/stable/io/fits/index.html

```python
#using random image from X-Drive to get latitude, longitude of BPO from the file's information
image_lat = fits.getval("star_Light_Blue_30_secs_001.fits", 'SITELAT')
image_long = fits.getval("star_Light_Blue_30_secs_001.fits", 'SITELONG')
print(f"BPO (in deg): Latitude = {image_lat}, Longitude = {image_long}")
```

co-ordinate systems guide: https://docs.astropy.org/en/stable/coordinates/index.html

```python
# reference frame for alt-az co-ords, set here to the latitude/longitude co-ords of BPO, at 10:00PM on Sunday 12th March
time = Time('2023-3-12 22:00:00')
bath = EarthLocation(lat = 51.37284*u.deg, lon = -2.319218*u.deg, height = 190.0*u.m)
# *u.unit changes variable to a 'Quantity' that stores both a value and its units
```

```python
#converting ra-dec to alt-az co-ordinates
altazdeg = SkyCoord(data['ra'], data['dec'], frame='icrs', unit='deg') 
altazcatalog = altazdeg.transform_to(AltAz(obstime=time,location=bath))
#adding alt-az co-ords to table
data['AltAz'] = altazcatalog.to_string('decimal')
data[['az','alt']] = data.AltAz.str.split(expand=True)
data["az"] = pd.to_numeric(data["az"])
data["alt"] = pd.to_numeric(data["alt"])
data = data.drop(['AltAz'], axis = 1)
data.head()
```

### Positions of the variable stars

```python
#example of what SIMBAD results table looks like
uma_result_table = customSimbad.query_object("RV Uma")
uma_result_table
```

```python
#getting right ascension, declination co-ords from SIMBAD
cep_result_table = customSimbad.query_object("RZ Cep")
cas_result_table = customSimbad.query_object("RZ Cas")
print(f"Right Ascension in units h:m:s, Declination in units d:m:s\nRZ Cep: Right Ascension = {cep_result_table[0][1]}, Declination = {cep_result_table[0][2]}")
print(f"RV Uma: Right Ascension = {uma_result_table[0][1]}, Declination = {uma_result_table[0][2]}")
print(f"RZ Cas: Right Ascension = {cas_result_table[0][1]}, Declination = {cas_result_table[0][2]}")
```

```python
#co-ords for RZ Cep
cepcoords = SkyCoord('22h39m13.1768s', '+64d51m30.611s', frame='icrs')
cepaltaz = cepcoords.transform_to(AltAz(obstime=time,location=bath))
print(f"RZ Cep: Altitude = {cepaltaz.alt:.6}, Azimuthal = {cepaltaz.az:.6}")

#co-ords for RV Uma
umacoords = SkyCoord('13h33m18.0838s', '+53d59m14.598s', frame='icrs')
umaaltaz = umacoords.transform_to(AltAz(obstime=time,location=bath))
print(f"RV Uma: Altitude = {umaaltaz.alt:.6}, Azimuthal = {umaaltaz.az:.6}")

#co-ords for RZ Cas
cascoords = SkyCoord('02h48m55.5102s', '+69d38m03.441s', frame='icrs')
casaltaz = cascoords.transform_to(AltAz(obstime=time,location=bath))
print(f"RZ Cas: Altitude = {casaltaz.alt:.6}, Azimuthal = {casaltaz.az:.6}")
```

### Altitude, Azimuth graph of the variable stars and all the standards

```python
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(1,1,1)

# plotting minimum altitude
x = np.linspace(0, 360)
y = np.linspace(30, 30)
plt.plot(x, y, color='black', ls='--', label = 'Minimum altitude')

#standards
plt.scatter(data['az'], data['alt'], marker = 'x', c='black', s=200, alpha = 1, label = 'All possible standards')
#RV Uma
plt.scatter(umaaltaz.az / u.deg, umaaltaz.alt / u.deg, marker = '*', c='darkseagreen', s=400, label = 'RV Uma')
#/u.deg changes quantity back to a unitless variable
#RZ Cas
plt.scatter(casaltaz.az / u.deg, casaltaz.alt / u.deg, marker = '*', c='cornflowerblue', s=400, label = 'RZ Cas')
#RZ Cep
plt.scatter(cepaltaz.az / u.deg, cepaltaz.alt / u.deg, marker = '*', c='lightcoral', s=400, label = 'RZ Cep')

plt.xlabel('Azimuth /$^\circ$', fontsize = 19)
plt.ylabel('Altitude /$^\circ$', fontsize = 19)
plt.xticks(size = 17)
plt.yticks(size = 17)
plt.xlim(0, 360)
plt.ylim(0, 90)

plt.legend(loc = "upper right")

plt.show()
```

### Remove stars that do not meet the telescope's specifications

```python
#telescope cannot see below an alttitude of around 30, therefore remove all those stars
data.drop(data[data['alt'] <= 30].index, inplace = True)
## telescope cannot see magnitudes below 6 or above 12, therefore remove all those stars
data.drop(data[(data['b mag'] >= 12)].index, inplace = True)
data.drop(data[(data['b mag'] <= 6)].index, inplace = True)
```

### RZ Cep's standards

.seperation() guide: https://docs.astropy.org/en/stable/coordinates/matchsep.html

```python
#creates an array with the data table indices of all stars with a seperation of < 45 degrees
cepcatalog = SkyCoord(alt = data['alt']*u.deg, az = data['az']*u.deg, frame='altaz', obstime=time, location=bath)
cep_d2d = cepaltaz.separation(cepcatalog)
cepcatalog_idx = np.where(cep_d2d < 50*u.deg)[0]
cepcatalog_idx
```

Changing the time will change the alt-az co-ords so max seperation will be different, can just change max seperation number around until you get a good number of stars in the table if the time needs to be changed.

```python
cep_standards = pd.DataFrame(columns=['Identifier', 'ra', 'dec', 'b mag', 'az', 'alt', 'seperation'])
#adding all the stars with their data table index listed in cepcatalog_idx to new array cep_standards
for i in range(cepcatalog_idx.shape[0]):
    cep_standards.loc[i] = data.iloc[cepcatalog_idx[i]]
#add the value for the seperation in arcseconds for each star into the table from the array of quantities cep_d2d
for i in range (cepcatalog_idx.shape[0]):
    cep_standards['seperation'].loc[i] = cep_d2d.arcsecond[cepcatalog_idx[i]]
# rank into order of increasing seperation
cep_standards = cep_standards.sort_values(by='seperation', ascending = True)
cep_standards
```

```python
#telescope cannot resolve stars with an angular seperation of less than 20 arcseconds

#replace the row index numbers in the table so that they're in order
cep_idx_list = []
for i in range (0, cep_standards.shape[0]):
    cep_idx_list.append(i)
cep_standards.index = cep_idx_list
#create 2d array that lists the indices of the pairs of stars which are too close together
cep_close_sep = []
for i in range (0, 8):
    cep_c1 = SkyCoord(alt = cep_standards['alt'].loc[i]*u.deg, az = cep_standards['az'].loc[i]*u.deg, frame='altaz', obstime=time, location=bath)
    cep_c2 = SkyCoord(alt = cep_standards['alt'].loc[i+1]*u.deg, az = cep_standards['az'].loc[i+1]*u.deg, frame='altaz', obstime=time, location=bath)
    cep_sep = cep_c1.separation(cep_c2)
    if cep_sep.arcsecond <= 20:
             cep_close_sep.append([i, i+1])
print(f"Indices of the pairs of stars that cannot be resolved: {cep_close_sep}")
```

```python
#For 12th March 2023, 22:00:00, no stars are too close therefore this section can be hashtagged out
#if there are stars that need to be removed, indicate their indices in labels to drop from table
#cep_standards.drop(labels = {}, axis = 0, inplace = True)
cep_standards
```

```python
#either, write first 5 rows into the final table:
#cepstandards_final = cep_standards.iloc[0:5]

#or, if all the stars have similar co-ords, preference is to have stars of mag 7-11 so just pick 5 of those out the table:
cepstandards_final = pd.DataFrame(columns=['Identifier', 'ra', 'dec', 'b mag', 'az', 'alt', 'seperation'])
cas_index = [0,1,3,5,7]
for i in cas_index:
    cepstandards_final = cepstandards_final.append(cep_standards.iloc[i])

#write into csv file for final list of standard stars
cepstandards_final.to_csv('rz_cep_standards.csv', header=True, index=False)
cepstandards_final
```
